```python
# These two functions will strip the dollar sign off of a string and replace commas within a number 1000 or greater.
# The first function will also remove the last 3 characters in the string which should be '.00' so they can be integers.
def dollar_to_int(row):
    return row.strip('$')[:-3]
def no_comma(row):
    return row.replace(',','')
```

```python
# There are NaN values within the 'security_deposit' column so, I made an empty list to hold row values obtained from a
# for loopreplacing the comma with no blank space ('').  The second 
security = []
security_n = []
for i in listings_summary['security_deposit'].dropna():
    security.append(i.replace(',',''))
for i in security:
    security_n.append(int(i.replace('$', '')[:-3]))
```

```python
cleaning = []
cleaning_fee = []
for i in listings_summary['cleaning_fee'].dropna():
    cleaning.append(i.replace(',',''))
for i in cleaning:
    cleaning_fee.append(int(i.replace('$', '')[:-3]))
cleaning_fee_mean = int(round(np.array(cleaning_fee).mean())) # 27
```

```python
security_dep_mean = int(round(np.array(security_n).mean())) # 202
```

```python
listings_summary['cleaning_fee'] = listings_summary['cleaning_fee'].fillna('$27.00')
```

```python
listings_summary['security_deposit'] = listings_summary['security_deposit'].fillna('$202.00')
```

```python
listings_summary['price'] = listings_summary['price'].apply(dollar_to_int).apply(no_comma).astype(int)
listings_summary['security_deposit'] = listings_summary['security_deposit'].apply(dollar_to_int).apply(no_comma).astype(int)
listings_summary['cleaning_fee'] = listings_summary['cleaning_fee'].apply(dollar_to_int).apply(no_comma).astype(int)
listings_summary['extra_people'] = listings_summary['extra_people'].apply(dollar_to_int).apply(no_comma).astype(int)
```
